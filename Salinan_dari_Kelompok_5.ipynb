#OUTLIER

```python
numerik = ['Harga Awal', 'Harga Setelah Diskon', 'Jumlah', 'Total Harga Produk', 'Total Diskon',
           'Diskon Dari Penjual', 'Diskon Dari Shopee', 'Jumlah Produk di Pesan', 'Potongan Koin Shopee',
           'Diskon Kartu Kredit', 'Ongkos Kirim Dibayar oleh Pembeli', 'Estimasi Potongan Biaya Pengiriman',
           'Ongkos Kirim Pengembalian Barang', 'Total Pembayaran', 'Perkiraan Ongkos Kirim']
for kolom in numerik:
    df[kolom] = pd.to_numeric(df[kolom], errors='coerce')

fig, axes = plt.subplots(nrows=len(numerik), figsize=(15, 6 * len(numerik)))

for i, kolom in enumerate(numerik):
    df.boxplot(column=[kolom], vert=False, ax=axes[i], widths=0.7, patch_artist=True, medianprops={'color':'black'})

    axes[i].set_title(f'Boxplot dari {kolom}', fontsize=14)
    axes[i].set_xlabel('Nilai', fontsize=12)
    axes[i].set_yticklabels([])

plt.tight_layout()
plt.suptitle('Distribusi Nilai untuk Setiap Kolom', fontsize=16, y=1.02)
import warnings
warnings.filterwarnings('ignore')
plt.show()
```

```python

def outliers(kolom):
    Q1 = np.percentile(df[kolom], 25, interpolation='midpoint')
    Q3 = np.percentile(df[kolom], 75, interpolation='midpoint')
    IQR = Q3 - Q1  # Perbaikan tanda minus

    # Upper bound
    upper = np.where(df[kolom] >= (Q3 + 1.5 * IQR))
    # Lower bound
    lower = np.where(df[kolom] <= (Q1 - 1.5 * IQR))

    print("Outliers upper bound:", df[kolom].iloc[upper])
    print("Outliers lower bound:", df[kolom].iloc[lower])

# Pemanggilan fungsi
outliers('Harga Awal')
```

dikarena dari data di ats menyatakan tidak banyak data yang hilang maka outlier bisa di ganti dengan memasukan nilai median dari kolom tersebut

```python


def impute_outliers_IQR(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    upper = series[series > (Q3 + 1.5 * IQR)].max()
    lower = series[series < (Q1 - 1.5 * IQR)].min()

    series = np.where(
        series > upper,
        series.median(),
        np.where(
            series < lower,
            series.median(),
            series
        )
    )
    return series
```

```python
# Memanggil fungsi dan menyimpan hasilnya kembali ke dalam dataframe
df['Harga Awal'] = impute_outliers_IQR(df['Harga Awal'])
df['Harga Setelah Diskon'] = impute_outliers_IQR(df['Harga Setelah Diskon'])
df['Jumlah'] = impute_outliers_IQR(df['Jumlah'])
df['Total Harga Produk'] = impute_outliers_IQR(df['Total Harga Produk'])
df['Total Diskon'] = impute_outliers_IQR(df['Total Diskon'])
df['Diskon Dari Penjual'] = impute_outliers_IQR(df['Diskon Dari Penjual'])
df['Jumlah Produk di Pesan'] = impute_outliers_IQR(df['Jumlah Produk di Pesan'])
df['Potongan Koin Shopee'] = impute_outliers_IQR(df['Potongan Koin Shopee'])
df['Ongkos Kirim Dibayar oleh Pembeli'] = impute_outliers_IQR(df['Ongkos Kirim Dibayar oleh Pembeli'])
df['Estimasi Potongan Biaya Pengiriman'] = impute_outliers_IQR(df['Estimasi Potongan Biaya Pengiriman'])
df['Ongkos Kirim Pengembalian Barang'] = impute_outliers_IQR(df['Ongkos Kirim Pengembalian Barang'])
df['Total Pembayaran'] = impute_outliers_IQR(df['Total Pembayaran'])
df['Perkiraan Ongkos Kirim'] = impute_outliers_IQR(df['Perkiraan Ongkos Kirim'])
```

```python
# reshaping dataset berdasarkan No Pesanan agar penggunaan lebih efektif dan cepat
data = df.groupby('No. Pesanan')['Nama Variasi'].unique()
data[:10]
```

```python
# cek total No pesanan
list_pesanan = data.tolist()
len(list_pesanan)
```

```python
# encoding produk pada sebuah list transaksi untuk kebutuhan apriori
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()
encode_data = encoder.fit(list_pesanan).transform(list_pesanan)
df_pesanan = pd.DataFrame(encode_data,columns=encoder.columns_)
df_pesanan
```

# Kombinasi Produk Menarik :



*   Memiliki asosiasi atau hubungan erat.
*   Kombinasi produk minimal 2 item, dan maksimum 3 item.
*   Kombinasi produk itu muncul setidaknya 10 dari dari seluruh transaksi.
*   Memiliki tingkat confidence minimal 50 persen.










```python
# setting pesanan berdasarkan kriteria menarik yang sudah di tuliskan diatas
df_pesanan = df_pesanan[df_pesanan.sum(axis=1) >= 2]
kriteria = 10 / len(list_pesanan)
kriteria
```

# Dalam kode di bawah:



*   frekuensi_itemset: Berisi hasil dari algoritma Apriori yang menemukan Frequent Itemsets berdasarkan persentase support.
*   frekuensi_sort: Mengurutkan Frequent Itemsets berdasarkan support secara menurun untuk menampilkan yang paling relevan atau signifikan terlebih dahulu.

**Frequent itemsets** adalah himpunan item-item yang sering muncul bersama dalam suatu dataset pesanan. Dalam konteks algoritma Apriori, frequent itemsets merujuk pada kombinasi item yang muncul dengan frekuensi di atas batas tertentu yang telah ditentukan (min_support). Frekuensi ini mengukur seberapa sering itemset tersebut muncul dalam dataset dibandingkan dengan jumlah total Pesanan.





```python
#implementasi algoritma apriori
frekuensi_itemset = apriori(df_pesanan, min_support=kriteria, max_len=3, use_colnames=True)
frekuensi_sort = frekuensi_itemset.sort_values(by='support', ascending=False).reset_index(drop=True)
frekuensi_sort
```

```python
# mendapatkan data asosiasi rules dengan metric lift
rules = association_rules(frekuensi_sort, metric = 'lift')
rules
```

```python
# pilih rules dengan nilai confindence 50% dan positif leverage
result = rules[(rules['confidence'] >= 0.5) & (rules['leverage'] > 0)] \
                .sort_values(by='lift', ascending=False)[:10]
result
```

Kombinasi **Taeyong** dan **Jaemin** terhadap **Douyoung** menghasilkan asosiasi atau hubungan yang paling erat dimana nilai confidence cukup tinggi diikuti metric lift yang paling tinggi.

```python
# Filter rules dengan confidence <= 0.2 untuk mengetahui item dengan asosiasi nilai terendah
low_confidence_rules = rules[rules['confidence'] <= 0.2]
most_frequent_consequents = low_confidence_rules['consequents'].value_counts().idxmax()
result_low_confidence = low_confidence_rules[low_confidence_rules['consequents'] == most_frequent_consequents]
# Filter rules dengan confidence >= 0.4 untuk mengetahui item dengan asosiasi nilai tertinggi
high_confidence_rules = rules[rules['confidence'] >= 0.4]
most_frequent_consequents = high_confidence_rules['consequents'].value_counts().idxmax()
result_high_confidence = high_confidence_rules[high_confidence_rules['consequents'] == most_frequent_consequents]

slow_moving = pd.concat([result_high_confidence, result_low_confidence])
slow_moving
```

*   Sedangkan variasi yang dipasangkan dengan jungwoo menghasilkan asosiasi atau hubungan yang tidak begitu erat dilihat dari nilai confidence yang rendah dan metric lift juga lebih rendah.
